```python
import requests
import matplotlib.pyplot as plt
import time


api_url = "http://api.open-notify.org/iss-now.json"
data_points = []
num_points = 100
for _ in range(num_points):
    response = requests.get(api_url)
    data = response.json()
    timestamp = int(data['timestamp'])
    latitude = float(data['iss_position']['latitude'])
    longitude = float(data['iss_position']['longitude'])
    data_points.append((latitude, longitude, timestamp))
    time.sleep(10)  


lats, lons, timestamps = zip(*data_points)


plt.scatter(lons, lats, c=timestamps, cmap='rainbow', alpha=0.7, s=20)
plt.colorbar(label='Timestamp')
plt.title("Path of the International Space Station")
plt.xlabel("Longitude")
plt.ylabel("Latitude")


plt.show()
```
